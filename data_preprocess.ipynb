```
def HistogramEqualization(pyIm):
    pyImNew = np.empty(pyIm.shape)

    num = int(np.max(pyIm.flatten())-np.min(pyIm.flatten()))+1
    im_hist, bins = np.histogram(pyIm.flatten(), num)

    cdf = im_hist.cumsum()
    cdf = max(pyIm.flatten()) * cdf /cdf[-1]

    image_equalized = np.interp(pyIm.flatten(),bins[:-1],cdf)
    pyImNew = np.reshape(image_equalized,pyIm.shape)
    
    return pyImNew
```

```
modality = ["ct", "mr"]
data_folder = '/content/gdrive/My Drive/ImageData/MMWHS'

for m in modality:
  hist_grams = []
  imgVol_fn, _ = getTrainNLabelNames(data_folder, m)
  plt.figure()
  fig, axs = plt.subplots(nrows=4, ncols=5, figsize=(12, 6), facecolor='w', edgecolor='k')
  for i, ax in enumerate(fig.axes):
    img_path = imgVol_fn[i]
    imgVol = sitk.GetArrayFromImage(sitk.ReadImage(img_path))  # numpy array
    imgVol = RescaleIntensity(imgVol, m, [750,-750])
    imgVol = HistogramEqualization(imgVol)
    imgVol = RescaleIntensity(imgVol, m, [np.max(imgVol),np.min(imgVol)])
    
    hist, bins = np.histogram(imgVol.flatten(), bins=50, density=True)
    width = 0.7 * (bins[1] - bins[0])
    center = (bins[:-1] + bins[1:]) / 2
    ax.bar(center, hist, align='center', width=width)
    ax.set_ylim(0,2.5)
    ax.set_yticklabels([])
  plt.show()
```
